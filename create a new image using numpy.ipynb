```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
photo = np.zeros((400,400,3), dtype="uint8")
```

```python
plt.imshow(photo)
```

```python
l = cv2.line(photo , (0,0), (400,400), (0,255,255))
plt.imshow(l)
```

```python
l2 = cv2.line(photo , (400,0), (0,400), (0,255,0), 5)
plt.imshow(l2)
```

```python
r = cv2.rectangle(photo, (0,0) ,(150,150) , (0,0,255), 2)
plt.imshow(r)
```
